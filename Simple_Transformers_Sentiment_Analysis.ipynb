```
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd
import logging
```

```
logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```
train_data = pd.read_csv('dataset_transformers.csv')
```

```
train_data
```

```
train_data = train_data.drop('Unnamed: 0', axis=1)
```

```
train_data.columns = ["text", "labels"]
```

```
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
train_data['labels'] = encoder.fit_transform(train_data['labels'])
```

```
X = train_data['text']
Y = train_data['labels']
```

```
X
```

```
Y
```

```
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
```

```
# prepare cross validation
n=5
kf = StratifiedKFold(n_splits=n, random_state=42, shuffle=True)
```

```
results = []
```

```
model_args = ClassificationArgs()

model_args.train_batch_size = 32
model_args.learning_rate = 1e-5
model_args.adam_epsilon = 1e-8
model_args.num_train_epochs = 10
model_args.overwrite_output_dir = True

import torch
cuda_available = torch.cuda.is_available()

model = ClassificationModel("roberta", "roberta-base", num_labels=3, args=model_args, use_cuda=cuda_available)
```

```
for train_index, val_index in kf.split(X,Y):
  train_df = train_data.iloc[train_index]
  val_df = train_data.iloc[val_index]

  model.train_model(train_df)

  result, model_outputs, wrong_predictions = model.eval_model(val_df, acc=accuracy_score)
  print(result['acc'])

  results.append(result['acc'])
```

```
print("results",results)
```

```
print(f"Mean-Precision: {sum(results) / len(results)}")
```

```
import numpy as np

print('Std for accuracy: ', np.std(results, axis=0))
```
